```python
def compute_gap(data):
    data['gap'] = data['stock_chg'].values - data['index_chg'].values
    return data

def create_newframe(old, stock_stand):
    new = pd.merge(old, stock_stand, on='trade_date',how='left')
    new = new.rename(columns={'ts_code_x':'stock_code', 'ts_code_y':'index_code',
                            'pct_chg_x':'stock_chg','pct_chg_y':'index_chg'})
    new = pd.merge(new, stock_info, left_on='stock_code', right_on = 'ts_code',how='left')
    new[['gap']] = ''
    new = compute_gap(new)
    return new.drop(['symbol','ts_code','index_code'],axis=1)
    

for i in dir_list:
    old_f = pd.read_csv("data/" + i, index_col=0)
    stock_stand = pd.DataFrame()
    if re.match("30", i):
        stock_stand = cy_index
    elif re.match("68", i):
        stock_stand = kc50_index
    elif re.match("60", i):
        stock_stand = sh_index
    elif re.match("00", i):
        stock_stand = sz_index
    else:
        print("error when" + index)
        continue
    new_f = create_newframe(old_f, stock_stand)
    new_f.to_csv("ndata/" + i)

print('end')
```
