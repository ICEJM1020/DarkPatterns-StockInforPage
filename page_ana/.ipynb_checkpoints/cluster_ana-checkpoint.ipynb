```python
EXCLUDE = ['data.eastmoney.com', 'v.163.com/static','mail','biz.ifeng.com/','comment.tie.163.com',
          'keyword=','quote.eastmoney.com/','baike.eastmoney.com/','v.ifeng.com','news.ifeng.com',
          'gentie.ifeng.com','finance.sina.cn/app','about','topic.eastmoney.com']
CLASS = [45,49,50,51,54,55,56,57,62,63,65,66,67,68,69,70,71,73,74,77,78,79,80,81,84,85,86,87,88,92,93,
         95,96,97,98,99,101,102,103,104,105,107,108,109,110,111,112,113,115,116,117,118,119,120,121,123,124,
        130,148,149,154,155,165,166,167,169,174,175,178,187,191,193,194,228,229]
SWING = ['jigou','mingjia','renwu','guba','blog']


dp_links_with_swing = []


for index, row in cluster_urls.iterrows():
    if_clude = False
    for i in EXCLUDE:
        if re.search(i,row['url']):
            if_clude = True
    
    if if_clude:
        continue
    
    if row['pca_cluster_hdbscan'] in CLASS:
        continue
        
    if row['url'] in dp_links_with_swing:
        continue
        
    dp_links_with_swing.append(row['url'])
    
print('finish')
```

```python
print(len(dp_links_with_swing))
pd.DataFrame(dp_links_with_swing).to_csv('dp_links_with_swing.csv',index = False)
pd.DataFrame(dp_links_with_swing).head(10)
```

```python
dp_links = []
for row in dp_links_with_swing:
    if_clude = False
    for i in SWING:
        if re.search(i,row):
            if_clude = True
            
    if if_clude:
        continue
        
    dp_links.append(row)
```

```python
print(len(dp_links))
pd.DataFrame(dp_links).to_csv('dp_links.csv',index = False)
pd.DataFrame(dp_links).head(10)
```
