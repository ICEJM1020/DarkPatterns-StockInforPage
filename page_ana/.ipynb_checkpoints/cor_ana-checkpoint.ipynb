```python
sz_index = ts.pro_bar(ts_code='399001.SZ', asset='I', start_date='20210315', end_date='20210409')[['ts_code','trade_date','pct_chg']]
sh_index = ts.pro_bar(ts_code='000001.SH', asset='I', start_date='20210315', end_date='20210409')[['ts_code','trade_date','pct_chg']]
kc50_index = ts.pro_bar(ts_code='000688.SH', asset='I', start_date='20210315', end_date='20210409')[['ts_code','trade_date','pct_chg']]
cy_index = ts.pro_bar(ts_code='399006.SZ', asset='I', start_date='20210315', end_date='20210409')[['ts_code','trade_date','pct_chg']]

def compute_gap(data):
    data['gap'] = data['stock_chg'].values - data['index_chg'].values
    return data

def create_newframe(old, stock_stand):
    new = pd.merge(old, stock_stand, on='trade_date',how='left')
    new = new.rename(columns={'ts_code_x':'stock_code', 'ts_code_y':'index_code',
                            'pct_chg_x':'stock_chg','pct_chg_y':'index_chg'})
    new = pd.merge(new, stock_info, left_on='stock_code', right_on = 'ts_code',how='left')
    new[['gap']] = ''
    new = compute_gap(new)
    return new.drop(['symbol','ts_code','index_code'],axis=1)

def getlistnum(li):
    li = list(li)
    set1 = set(li)
    dict1 = {}
    for item in set1:
        dict1.update({item:li.count(item)})
    return dict1
```

```python
men_ab_data = []
for index, row in data.iterrows():
    temp_sym = row['ts_code']
    stock_data = ts.pro_bar(
            ts_code=temp_sym, adj='qfq', 
            start_date='20210315', end_date='20210409', factors=['tor','vr'])
    
    stock_stand = pd.DataFrame()
    if re.match("30", temp_sym):
        stock_stand = cy_index
    elif re.match("68", temp_sym):
        stock_stand = kc50_index
    elif re.match("60", temp_sym):
        stock_stand = sh_index
    elif re.match("00", temp_sym):
        stock_stand = sz_index
    else:
        print("error when" + index)
        continue
        
    for j,jr in stock_data.iterrows():
        if jr.isnull().values.any():
            stock_data = stock_data.drop(j)
    
    new_f = create_newframe(stock_data, stock_stand)
    new_f = new_f.drop(['trade_date','name','stock_code'],axis=1)
    
    ans = clf.predict(new_f)
    row['ab_counts'] = list(ans).count(1)
    
    men_ab_data.append(row)
```

```python
from matplotlib.pyplot import MultipleLocator
import matplotlib.pyplot as plt
```

```python
def compute_aver(inpute_data):
    test_data = {}
    for index, row in inpute_data.iterrows():
        if row['counts'] in test_data.keys():
            test_data[row['counts']][0] += row['ab_counts']
            test_data[row['counts']][1] += 1
        else:
            test_data[row['counts']] = [row['ab_counts'],1]
    return test_data

aver = compute_aver(men_ab_data)
```

```python
def compute_pro(inpute_data):
    test_data = {}
    for index, row in inpute_data.iterrows():
        if row['counts'] in test_data.keys():
            test_data[row['counts']][0] += row['if_ab']
            test_data[row['counts']][1] += 1
        else:
            test_data[row['counts']] = [row['if_ab'],1]
    return test_data

prob = compute_pro(men_ab_data)
```
